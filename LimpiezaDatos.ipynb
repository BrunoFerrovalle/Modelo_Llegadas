Iniciamos con la importacion basicas de librerias para el procesado y manejo de la informacion

```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

Cargamos el dataset de calculo y el principal

```python
ModelData = pd.DataFrame(pd.read_csv("Tcargos.csv"))

Car_len_mean = pd.DataFrame(pd.read_csv("Tcargos.csv"))
Car_len_mean.head()
```

Procedemos a encontrar la media del largo de los coches para homologar el valor en el dataset


```python
def clean_data(Car_len_mean):
    # Performed 1 aggregation grouped on column: 'tipoCarro'
    Car_len_mean = Car_len_mean.groupby(['tipoCarro']).agg(carLength_mean=('carLength', 'mean')).reset_index()
    return Car_len_mean

Car_len_mean = clean_data(Car_len_mean.copy())
Car_len_mean.head()
```

Remplazamos el valor de carLength las filas donde el tipo de coche coincide para homologar los datos

```python
# Realizar el merge entre Tcargos y CarLength_Mean
merged_df = ModelData.merge(Car_len_mean, on='tipoCarro', how='left')

# Actualizar la columna carLength con los valores de carLength_mean
merged_df['carLength'] = merged_df['carLength_mean']

# Eliminar la columna carLength_mean que ya no necesitamos
merged_df.drop(columns=['carLength_mean'], inplace=True)
```

De igual manera con consignatario, tipoServicio, idTrenICHR,idFerrocarilICHR

```python

# Crear diccionarios que asignen un ID único a cada valor de las columnas especificadas
consignatario_id_map = {value: idx for idx, value in enumerate(merged_df['consignatario'].unique())}
tipoServicio_id_map = {value: idx for idx, value in enumerate(merged_df['tipoServicio'].unique())}
idTrenICHR_id_map = {value: idx for idx, value in enumerate(merged_df['idTrenICHR'].unique())}
idFerrocarrilICHR_id_map = {value: idx for idx, value in enumerate(merged_df['idFerrocarrilICHR'].unique())}
tipoCarro_id_map = {value: idx for idx, value in enumerate(merged_df['tipoCarro'].unique())}


# Mapear los valores de las columnas a sus correspondientes IDs
merged_df['consignatario_id'] = merged_df['consignatario'].map(consignatario_id_map)
merged_df['tipoServicio_id'] = merged_df['tipoServicio'].map(tipoServicio_id_map)
merged_df['idTrenICHR_id'] = merged_df['idTrenICHR'].map(idTrenICHR_id_map)
merged_df['idFerrocarrilICHR_id'] = merged_df['idFerrocarrilICHR'].map(idFerrocarrilICHR_id_map)
merged_df['tipoCarro_id'] = merged_df['tipoCarro'].map(tipoCarro_id_map)


# Crear DataFrames a partir de los diccionarios de mapeo
consignatario_id_df = pd.DataFrame(list(consignatario_id_map.items()), columns=['Consignatario', 'id'])
tipoServicio_id_df = pd.DataFrame(list(tipoServicio_id_map.items()), columns=['tipoServicio', 'id'])
idTrenICHR_id_df = pd.DataFrame(list(idTrenICHR_id_map.items()), columns=['idTrenICHR', 'id'])
idFerrocarrilICHR_id_df = pd.DataFrame(list(idFerrocarrilICHR_id_map.items()), columns=['idFerrocarrilICHR', 'id'])
tipoCarro_id_df = pd.DataFrame(list(tipoCarro_id_map.items()), columns=['tipoCarro', 'id'])


# Guardar los DataFrames de mapeo en archivos CSV
consignatario_id_df.to_csv('consignatario_id_map.csv', index=False)
tipoServicio_id_df.to_csv('tipoServicio_id_map.csv', index=False)
idTrenICHR_id_df.to_csv('idTrenICHR_id_map.csv', index=False)
idFerrocarrilICHR_id_df.to_csv('idFerrocarrilICHR_id_map.csv', index=False)
tipoServicio_id_df.to_csv('tipoCarro_id_map.csv', index=False)


# Eliminar las columnas originales
merged_df.drop(columns=['consignatario', 'tipoServicio', 'idTrenICHR', 'idFerrocarrilICHR',"tipoCarro"], inplace=True)

# Mostrar las primeras filas del DataFrame actualizado
merged_df.head()
```

Ahora para hacer discreto el campo de fechaEvento se va a separar entre 3 columnas donde a continuacion solo se pondran la de añoMes, dia para posteriormente mapear y poner el campo de hora

```python
merged_df['fechaEvento'] = pd.to_datetime(merged_df['fechaEvento'])
# Extraer el año, mes y día
merged_df['año'] = merged_df['fechaEvento'].dt.year
merged_df['mes'] = merged_df['fechaEvento'].dt.month
merged_df['día'] = merged_df['fechaEvento'].dt.day

# Crear un nuevo campo 'añomes' concatenando el año y el mes
merged_df['añoMes'] = (merged_df['año'] * 100 + merged_df['mes']).astype(int)

# Crear una columna de fecha en merged_df
merged_df['fecha'] = pd.to_datetime(merged_df['añoMes'].astype(str) + merged_df['día'].astype(str), format='%Y%m%d')

# Calcular el número de coches por día
coches_por_dia_merged = merged_df.groupby('fecha').size().reset_index(name='coches_por_dia')

# Unir la columna coches_por_dia a merged_df
merged_df = merged_df.merge(coches_por_dia_merged, on='fecha', how='left')
merged_df.head()

# Eliminar las columnas temporales de año y mes si no se necesitan
merged_df.drop(columns=['año', 'mes','fechaEvento','fecha'], inplace=True)

# Mostrar las primeras filas del DataFrame actualizado
merged_df.head()
```

Se procede con el campo hora para tenerlo en formato entero

```python
# Asegurarse de que la columna 'hora' esté en el formato datetime
merged_df['hora'] = pd.to_datetime(merged_df['hora'], format='%H:%M:%S')

# Redondear las horas al intervalo de 5 minutos más cercano
def round_time_to_nearest_five_minute(t):
    discard = pd.Timedelta(minutes=t.minute % 5, seconds=t.second, microseconds=t.microsecond)
    return t - discard + (pd.Timedelta(minutes=5) if discard >= pd.Timedelta(minutes=2.5) else pd.Timedelta(0))

merged_df['hora'] = merged_df['hora'].apply(round_time_to_nearest_five_minute)

# Extraer la hora y minuto de la hora redondeada para mantenerlas en formato de tiempo
merged_df['hora'] = merged_df['hora'].dt.time

# Extraer la hora, minuto y segundo
merged_df['hora_h'] = merged_df['hora'].apply(lambda x: x.hour)
merged_df['hora_m'] = merged_df['hora'].apply(lambda x: x.minute)
merged_df['hora_s'] = merged_df['hora'].apply(lambda x: x.second)

merged_df['hora_minuto'] = (merged_df['hora_h'].astype(str).str.zfill(2) + merged_df['hora_m'].astype(str).str.zfill(2)).astype(int)

merged_df.drop(columns=['hora_h', 'hora_m', 'hora_s','hora'], inplace=True)

```
